### Visualising a Simple Tree

```python
from Bio import Phylo
tree = Phylo.read("simple.dnd", "newick")
```

```python
print(tree)
```

```python
Phylo.draw_ascii(tree)
```

```python
# Draw the tree graphically (requires Matplotlib or PyLab)
tree.rooted = True
Phylo.draw(tree)
```

### Coloring Tree Branches

```python
from Bio.Phylo.PhyloXML import Phylogeny
tree = tree.as_phyloxml()
tree = Phylogeny.from_tree(tree)
```

```python
# Coloring the whole tree gray by assigning root color
# Colors for a clade are treated as cascading down through the entire clade, 
# thus assigning the root colors the whole tree
# Also possible to enter RGB value (128,128,128) or hex #808080
tree.root.color = "gray" 

# Most recent commont ancestor
mrca = tree.common_ancestor({"name": "E"}, {"name": "F"})
mrca.color = "salmon"

# Coloring by the index of the clade
tree.clade[0, 1].color = "blue"
```

```python
Phylo.draw(tree)
```

```python
# Save the colored tree to an xml file
Phylo.write(tree, "colored_simple_tree.xml", "phyloxml")
```
